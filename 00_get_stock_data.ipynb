```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bist30h = [
           'TUPRS.IS', 
           'VAKBN.IS',
           'KRDMD.IS',
           'DOHOL.IS',
           'AKBNK.IS',
           'TKFEN.IS',
           'FROTO.IS',
           'GARAN.IS',
           'KOZAA.IS',
           'HALKB.IS',
           'TCELL.IS',
           'KOZAL.IS',
           'VESTL.IS',
           'YKBNK.IS',
           'TTKOM.IS',
           'BIMAS.IS',
           'PETKM.IS',
           'SISE.IS',
           'ARCLK.IS',
           'EREGL.IS',
           'SASA.IS',
           'SAHOL.IS',
           'EKGYO.IS',
           'THYAO.IS',
           'PGSUS.IS',
           'ISCTR.IS',
           'KCHOL.IS',
           'ASELS.IS',
           'TAVHL.IS',
           'GUBRF.IS',
]
```

```python
data = yf.download(
    tickers = ' '.join(bist30h),
    period = "max",
    start = "2005-01-01"
)

data.clip(lower=0, inplace=True)
data.fillna(0, inplace=True)
```

```python
data['Adj Close'].plot()
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.tight_layout(pad=7)
plt.show()
```

```python
data.to_csv("00_data.csv")
```
